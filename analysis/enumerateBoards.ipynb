```python
import numpy as np
import copy as copy

boards = [];

ship_sizes = [3,2,2]; 
n = 5; #number or rows
m = 5; #number of columns

def placeShip(ship_idx,board_state,boards):
    
    ship_size=ship_sizes[ship_idx]
    for i in range(0,n):
        for j in range(0,m):
            
            #is it possible to go down?
            if ((i+ship_size<=n) and  #if there is enough space
            (np.all(board_state[i:i+ship_size,j]==0)) and  #and we are not overlapping with any ship
            ((i==0) or (board_state[i-1,j]==0)) and  # and there is no ship right above
            ((i+ship_size==n) or (board_state[i+ship_size,j]==0)) and  # and there is no ship right below
            ((j==0) or (np.all(board_state[i:i+ship_size,j-1]==0))) and  #and there are no ships touching on the left
            ((j==m-1) or (np.all(board_state[i:i+ship_size,j+1]==0)))): # and there are no ships touching on the right
                proposed_board_state = copy.deepcopy(board_state);
                proposed_board_state[i:i+ship_size,j]=1;
                if ship_idx==len(ship_sizes)-1:
                    boards.append(proposed_board_state)
                else:
                    boards = placeShip(ship_idx+1,proposed_board_state,boards)
#                 print((i,j, 'down'))
    
            #is it possible to go right?
            if ((j+ship_size<=m) and  #if there is enough space
                (np.all(board_state[i,j:j+ship_size]==0)) and  #and we are not overlapping with any ship
                ((j==0) or (board_state[i,j-1]==0)) and  # and there is no ship to the left
                ((j+ship_size==m) or (board_state[i,j+ship_size]==0)) and  # and there is no ship to the right
                ((i==0) or (np.all(board_state[i-1,j:j+ship_size]==0))) and  #and there are no ships touching on top
                ((i==n-1) or (np.all(board_state[i+1,j:j+ship_size]==0)))): # and there are no ships touching below
                    proposed_board_state = copy.deepcopy(board_state);
                    proposed_board_state[i,j:j+ship_size]=1;
                    if ship_idx==len(ship_sizes)-1:
                        boards.append(proposed_board_state)
                    else:
                        boards = placeShip(ship_idx+1,proposed_board_state,boards)
                    
    return(boards)


    

                
```

```python
boards = [];

ship_sizes = [3,2,2]; 
n = 5; #number or rows
m = 5; #number of columns

ship_idx=0
board_state = np.empty((n,m));
board_state[:] = 0;
boards=placeShip(ship_idx,board_state,[])
```

```python
boards_array = np.array(boards)
```

```python
np.mean(boards_array,0)
```

```python
flat_boards = np.array([np.reshape(b,25) for b in boards]);
np.savetxt("flat_boards.csv", flat_boards, delimiter=",");

```
